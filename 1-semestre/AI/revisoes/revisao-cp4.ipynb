```python
import pandas as pd
```

```python
df_modificado = pd.read_csv('../datasets/misterio_trem_modificado.csv')
df_original = pd.read_csv('../datasets/misterio_trem_original.csv')
```

```python
df_vitima = df_original[~df_original['ID'].isin(df_modificado['ID'])]
df_vitima
```

```python
hora_saida_vitima = df_vitima['Hora_Saída_Restaurante'].values[0]
hora_entrada_vitima = df_vitima['Hora_Entrada_Restaurante'].values[0]
hora_saida_vitima
```

```python
df_presentes_no_restaurante = df_modificado[
    (df_modificado['Hora_Entrada_Restaurante'] == hora_entrada_vitima)
    & (df_modificado['Hora_Saída_Restaurante'] == hora_saida_vitima)
]
df_presentes_no_restaurante
```

```python
df_objetos_unicos = df_modificado['Objeto_Carregado'].unique()
df_objetos_unicos
```

```python
objetos_perigosos = ['Bastão de Camiinhada', 'Faca', 'Bengala']

df_suspeitos = df_presentes_no_restaurante[df_presentes_no_restaurante['Objeto_Carregado'].isin(objetos_perigosos)]
df_suspeitos
```

```python
destino_vitima = df_vitima['Destino'].values[0]
df_suspeitos = df_suspeitos[df_suspeitos['Destino'] == destino_vitima]
df_suspeitos
```

```python
valor_gasto_vitima = df_vitima['Valor_Gasto'].values[0]
df_assassino = df_suspeitos[df_suspeitos['Valor_Gasto'] > valor_gasto_vitima]
df_assassino
```

```python
objetos_inofensivos = [obj for obj in df_objetos_unicos if obj not in objetos_perigosos]

df_testemunhas = df_presentes_no_restaurante[df_presentes_no_restaurante['Objeto_Carregado'].isin(objetos_inofensivos)]
df_testemunhas_ordenadas = df_testemunhas.sort_values('Nome')
df_testemunhas_ordenadas
```
